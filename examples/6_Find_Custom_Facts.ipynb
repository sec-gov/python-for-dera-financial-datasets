### Read Submissions, Nemeric and Tags Data Sets for Feb. 2024 from [Financial Statement and Notes Data Sets](https://www.sec.gov/dera/data/financial-statement-and-notes-data-set)

```python
# Setting low_memory parameter to False to prevent mixed type inference and default_na parameter to False so that empty strings are not parsed as NaN.
submissions = pd.read_csv('data/2024_02_notes/sub.tsv', sep='\t', low_memory=False, keep_default_na=False)
numericFacts = pd.read_csv('data/2024_02_notes/num.tsv', sep='\t', low_memory=False, keep_default_na=False)
tags = pd.read_csv('data/2024_02_notes/tag.tsv', sep='\t', low_memory=False, keep_default_na=False)
```

### Filter for Custom Monetary Tags

```python
# use value of abstarct column as 0 to filter for concrete (non-abstract) tags
customTagsDF = tags[(tags.custom == 1) & (tags.abstract == 0) & (tags.datatype == 'monetary')]
customTagsDF
```

### Merge submissions Data Frame with numericFacts Data Frame Using adsh Column as a Join Key

```python
mergedSubmissionsAndNumericFactsDF = pd.merge(submissions, numericFacts, on='adsh')
```

### Merge mergedSubmissionsAndNumericFactsDF Data Frame with customTagsDF Data Frame Using adsh and tag as Join Keys

```python
# Please note that version column = adsh for custom tags
mergedDF = pd.merge(mergedSubmissionsAndNumericFactsDF, customTagsDF, left_on=['adsh', 'tag'], right_on=['version', 'tag'])
```

### Look at the Contents of mergedDF Data Frame

```python
mergedDF
```

### Filter for Facts with Custom Tags for AMAZON COM INC

```python
# Using data date (ddate) = 20231231 for year-end facts and dimn = 0 for non-dimensional facts
filter = (mergedDF.name == 'AMAZON COM INC') & (mergedDF.ddate == 20231231) & (mergedDF.dimn == 0)
# Sort values by tag column
customFactsForAmazon = mergedDF[filter][['name', 'tag', 'dimn', 'ddate', 'qtrs', 'uom', 'value']].sort_values(by= 'tag')
customFactsForAmazon.reset_index(drop=True, inplace=True)
customFactsForAmazon
```
