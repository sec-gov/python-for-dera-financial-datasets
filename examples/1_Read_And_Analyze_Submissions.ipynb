### Create a New DataFrame with a Subset of Columns

```python
subsetDF = filingsQ1[['name', 'countryinc', 'afs', 'form', 'period', 'fy', 'fp', 'filed']]
subsetDF
```

### Group Submissions By Form Column and Compute Group Sizes

```python
formGroupBy = subsetDF.groupby('form')
formGroups = formGroupBy.size().reset_index(name='count')
# Sort the results by count
formGroups = formGroups.sort_values(by = ['count'])
formGroups
```

### Filter Submissions for Important Forms (Form 10-K for Annual Filings, Form 10-Q forQuarterly Filings and  Form 8-K for Current Reports)

```python
importantForms = ['10-K', '10-K/A', '10-Q', '10-Q/A', '8-K', '8-K/A']
filter = formGroups['form'].isin(importantForms)
importantFormGroups = formGroups[filter]
importantFormGroups
```

### Visualize Statistics for Important Forms using Pyplot Library

```python
# Create a bar chart with form type on x axis and counts on y axis
importantFormGroups.plot(x='form', y='count', kind='bar', figsize=(8,6))
# Set title and labels for x & y axes
plt.title('Structured Data Form Statistics for 2022 Q1',fontsize=15)
plt.xlabel('Form Type', fontsize=15)
plt.ylabel('Number of Filings', fontsize=15)
plt.yticks(np.arange(0, 18001, step=1000))
plt.xticks(rotation=0)
# Save the bar chart as a file in images folder
plt.savefig('images/FormStats2022Q1.jpg')
plt.show()
```
