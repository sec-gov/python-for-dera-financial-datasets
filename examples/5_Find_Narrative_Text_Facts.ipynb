<h1 align="center">Join Submission and Text Data Sets to Find Narrative Facts</h1>

### Import Code Libraries

```python
import pandas as pd
import numpy as np
from IPython.display import display, HTML
# Set the pandas column width to unlimited
pd.set_option('display.max_colwidth', None)
# Set the notebook width to 100%
display(HTML("<style>.container { width:100% !important; }</style>"))
```

### Read Submissions and Text Data Sets for Feb. 2024 from [Financial Statement and Notes Data Sets](https://www.sec.gov/dera/data/financial-statement-and-notes-data-set)

```python
# Setting low_memory parameter to False to prevent mixed type inference and default_na parameter to False so that empty strings are not parsed as NaN.
submissions = pd.read_csv('data/2024_02_notes/sub.tsv', sep='\t', low_memory=False, keep_default_na=False)
textFacts = pd.read_csv('data/2024_02_notes/txt.tsv', sep='\t', low_memory=False, keep_default_na=False)
```

### Look at the Contents of submissions Data Frame

```python
submissions
```

### Look at the Contents of textFacts Data Frame

```python
textFacts
```

### Merge submissions Data Frame with textFacts Data Frame Using adsh Column as a Join Key

```python
mergedDF = pd.merge(submissions, textFacts, on='adsh')
```

### Look at the Contents of mergedDF Data Frame

```python
mergedDF
```

### Filter for Significant Accounting Policies for Marriott International using SignificantAccountingPoliciesTextBlock tag

```python
filter = (mergedDF.name == 'MARRIOTT INTERNATIONAL INC /MD/') & (mergedDF.tag == 'SignificantAccountingPoliciesTextBlock') 
# create a data frame with a subset of columns
textFactsForMarriott = mergedDF[filter][['name', 'tag', 'value']]
# reset the index of textFactsForMarriott data frame
textFactsForMarriott.reset_index(drop=True, inplace=True)
#left align the columns and headers
leftAlignedDF = textFactsForMarriott.style.set_properties(**{'text-align': 'left', 'vertical-align':'top'})
leftAlignedDF.set_table_styles([dict(selector='th', props=[('text-align', 'left')])])
```

### Filter for Exchange Act Section 12(g) Securities 

```python
filterFor12gSecurities = mergedDF.tag == 'Security12gTitle'
factsFor12gSecurities = mergedDF[filterFor12gSecurities][['name', 'cik', 'form', 'period', 'fy', 'fp', 'filed','tag', 'value']]
# Sort the factsFor12gSecurities data frame  by name column followed by value column
factsFor12gSecurities = factsFor12gSecurities.sort_values(by = ['name', 'value'])
factsFor12gSecurities.reset_index(drop=True, inplace=True)
factsFor12gSecurities
```

### Create a Excel file for Section 12(g) Securities

```python
# Save the Excel file in data folder
factsFor12gSecurities.to_excel('data\Section12gSecurities.xlsx', index=False)
```
